```python
from tensorflow.keras.applications.resnet50 import ResNet50
#ResNet-50 is a 50-layer convolutional neural network (48 convolutional layers, one MaxPool layer,
#and one average pool layer)
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
#to operate by utilizing the pre-trained weights from the ImageNet-trained model. 
#it's possible to achieve more successful results without wasting time on backpropagation processes 
```

```python
img_path = 'tennisboll.jpg' 
img = image.load_img(img_path, target_size=(224,224))
img
```

```python
x = image.img_to_array(img)
x.shape
# 3 dimesion 
```

```python
x = np.expand_dims(x, axis=0)
# used to add the number of images
x.shape
# 1 = n is the number of images process 
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)

#onvert the input images from RGB to BGR, 
#then will zero-center each color channel with respect to the ImageNet dataset,
# without scaling
```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=1)[0]) # top 3 or top 2 
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
# synset name
```
